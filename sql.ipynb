```python
import pandasql as ps
import pandas as pd
df = pd.read_csv(r'C:\Big data\merged_file.csv')
df
```

```python
query = """
SELECT ประเทศ as country  
FROM df
ORDER BY country ASC
"""

result = ps.sqldf(query, locals())

result.head(10)
```

```python
query = """
SELECT 
    เดือน,
    ประเทศ,
    การขนส่ง,
    SUM(ปริมาณ) AS Total_Quantity
FROM 
    df
GROUP BY 
    เดือน, ประเทศ, การขนส่ง
ORDER BY 
    Total_Quantity DESC
"""

result = ps.sqldf(query, locals())
result.head(5)
```
